**USE mtcars default data**

```r
install.packages("ggplot2")
library(ggplot2)
```

1.Histogram of hwy mpg values.

```r
ggplot(data=mtcars , aes(x=mpg ))+
  geom_histogram(binwidth = 3, fill = "pink", color = "black")+
  labs(x="MPG",y="Frequency",title = "Histogram of MPG Values in mtcars Dataset")
```

2.Barplot of carb counts  with color fill defined by cyl count.

```r
ggplot(data = mtcars, aes(x = factor(carb), fill = factor(cyl))) +
  geom_bar() +
  labs(x = "Carburetor Count", y = "Frequency", title = "Carburetor Counts") +
  scale_fill_manual(values = c("pink", "dark grey", "violet"))
```

3.Switch now to use the txhousing dataset that comes with ggplot2.Create a scatterplot of volume versus sales. Afterwards play around with alpha and color arguments to clarify information.


```r
# Sample data (random numbers)
set.seed(123)
txhousing_sample = txhousing[sample(nrow(txhousing), 1000), ]  # Sample 1000 rows from txhousing
txhousing_sample$volume = rnorm(1000, mean = 500, sd = 100)  # Volume (random numbers)
txhousing_sample$sales = rnorm(1000, mean = 1000, sd = 200)  # Sales (random numbers)

# Create a scatter plot of volume versus sales
ggplot(data = txhousing_sample, aes(x = volume, y = sales)) +
  geom_point(alpha = 0.6, color = "purple") +
  labs(x = "Volume", y = "Sales", title = "Volume vs Sales")
```

4.Add a smooth fit line to the scatterplot from above. Hint: You may need to look up geom_smooth().


```r
# Sample data (random numbers)
set.seed(123)
txhousing_sample = txhousing[sample(nrow(txhousing), 1000), ]  # Sample 1000 rows from txhousing
txhousing_sample$volume = rnorm(1000, mean = 500, sd = 100)  # Volume (random numbers)
txhousing_sample$sales = rnorm(1000, mean = 1000, sd = 200)  # Sales (random numbers)

# Create a scatter plot of volume versus sales
ggplot(data = txhousing_sample, aes(x = volume, y = sales)) +
  geom_point(alpha = 0.6, color = "purple") +
  geom_smooth(method="lm",col = "purple")+
  labs(x = "Volume", y = "Sales", title = "Volume vs Sales")
```

**USE airquality default data**

1.The x-axis represents the Day of the month.

2.The y-axis represents the Ozone levels.

3.Color the points by Month.

4.Add a smooth trend line to the plot.


```r
ggplot(data=airquality,aes(x=Day,y=Ozone,color=factor(Month)))+
  geom_point()+
  geom_smooth(method="lm",se=FALSE)+
  labs(x="Day of the month",y="Ozone levels",title = "Ozone Levels by Day of the Month")
```

**USE chickweight default data**

1.Plots Time on the x-axis and weight on the y-axis.

2.Uses different colors to represent each Diet.

3.Groups the lines by Chick to show individual growth trajectories within each diet.


```r
ggplot(data=ChickWeight ,aes(x=Time,y=weight,color=factor(Diet),group = Chick))+
  geom_line()+
  labs(x = "Time", y = "Weight")
```

**USE diamonds default data**


Create a histogram to show the distribution of carat sizes. Then, refine your plot by:

1.Filling the bars based on the cut quality of the diamonds.

2.Limiting the x-axis to diamonds up to 3 carats for better detail.


```r
ggplot(data = diamonds, aes(x = carat, fill = cut)) +
  geom_histogram(binwidth = 0.1, color = "black") +
  xlim(0, 3) +
  labs(x = "Carat", y = "Frequency", title = "Distribution of Carat Sizes by Cut Quality")
```
